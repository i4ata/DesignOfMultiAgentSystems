```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
import numpy as np
```

```python
all_data = pd.read_parquet('all_data.parquet')
mean_data = pd.read_parquet('averaged_data.parquet').set_index('[run number]').sort_index()
```

```python
sns.lineplot(data=all_data[all_data['[run number]'] == 5], x='[step]', y='voter-turnout')
plt.xticks(np.arange(0, 151, 15))
plt.xlabel('Step')
plt.ylabel('% Voter Turnout')
plt.title('Voter turnout in run #5')
plt.savefig('plots/run5.pdf')
plt.show()
```

```python
sns.lineplot(data=all_data, x='[step]', y='voter-turnout', hue='election-type', errorbar=None)
plt.legend(loc='lower right', title='election type')
plt.xticks(np.arange(0, 151, 15))
plt.xlabel('Step')
plt.ylabel('% Voter Turnout')
plt.title('Mean voter turnout per step')
plt.savefig('plots/mean_runs.pdf')
plt.show()
```

```python
sns.lineplot(data=all_data, x='[step]', y='voter-turnout', hue='election-type', errorbar='sd', err_kws={'alpha': .1})
plt.legend(loc='lower right', title='election type')
plt.xticks(np.arange(0, 151, 15))
plt.xlabel('Step')
plt.ylabel('% Voter Turnout')
plt.title('Mean voter turnout per step $\pm$ standard deviation')
plt.savefig('plots/mean_runs_with_std.pdf')
plt.show()
```

```python
mapping = dict(limited=0, plurality=1)
mean_data['election-type'] = mean_data['election-type'].map(mapping)
```

```python
X = mean_data.drop('voter-turnout', axis='columns')
X['intercept'] = 1
y = mean_data['voter-turnout']
```

```python
model = sm.OLS(y, X).fit()
```

```python
print(model.summary())
```

```python
sns.displot(model.resid)
plt.xlabel('Residual')
plt.title('Residual distribution')
plt.savefig('plots/dist_resid.pdf')
plt.show()
```

```python
sns.jointplot(x=model.predict(X), y=model.resid, kind='hex')
plt.xlabel('Predicted')
plt.ylabel('Residuals')
plt.axhline(0, color='red', ls='--')
plt.savefig('plots/residual_plot.pdf')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture

# Sample data composed of two normal distributions
data = np.concatenate([np.random.normal(30, 5, 5000), np.random.normal(70, 8, 500)])

# Fit a GMM to the data
gmm = GaussianMixture(n_components=2, random_state=0)
gmm.fit(data.reshape(-1, 1))

# Create a histogram of the data
plt.hist(data, bins=30, density=True, alpha=0.5, label='Data')

# Generate x-values for the PDF
x = np.linspace(min(data), max(data), 1000)

# Plot the scaled PDFs of the GMM components
for i in range(gmm.n_components):
    mean, std_dev = gmm.means_[i][0], np.sqrt(gmm.covariances_[i][0][0])
    weight = gmm.weights_[i]
    pdf = weight * np.exp(-(x - mean) ** 2 / (2 * std_dev ** 2)) / (std_dev * np.sqrt(2 * np.pi))
    plt.plot(x, pdf, label=f'Component {i + 1}')

plt.legend()
plt.title('Histogram with Overlaid GMM Components')
plt.show()
```
