```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.mixture import GaussianMixture
```

```python
mean_data = pd.read_parquet('averaged_data.parquet').set_index('[run number]').sort_index()
```

```python
sns.histplot(data=mean_data, x='voter-turnout', hue='election-type', stat='density')
```

```python
data = mean_data[mean_data['election-type'] == 'limited']['voter-turnout'].values.reshape(-1, 1)
```

```python
gmm = GaussianMixture(n_components=2, random_state=0)
gmm.fit(data)
preds = gmm.predict(data)
```

```python
mean_data['preds'] = 2
mean_data.loc[mean_data[mean_data['election-type'] == 'limited'].index, 'preds'] = preds
```

```python
mean_data['preds'].value_counts()
```

```python
mean_data.drop('election-type', axis=1).groupby('preds').agg(['mean', 'std'])
```

```python
from scipy.stats import ttest_ind
```

```python
x1 = mean_data[(mean_data['election-type'] == 'limited') & (mean_data['preds'] == 1)]['voter-turnout']
x3 = mean_data[(mean_data['election-type'] == 'limited') & (mean_data['preds'] == 0)]['voter-turnout']
x2 = mean_data[mean_data['election-type'] == 'plurality']['voter-turnout']
ttest_ind(x2, x3, equal_var=False)
```

```python
sns.histplot(data, stat='density')
x = np.linspace(min(data), max(data), 1000)

# Plot the scaled PDFs of the GMM components
for i in range(gmm.n_components):
    mean, std_dev = gmm.means_[i][0], np.sqrt(gmm.covariances_[i][0][0])
    weight = gmm.weights_[i]
    pdf = weight * np.exp(-(x - mean) ** 2 / (2 * std_dev ** 2)) / (std_dev * np.sqrt(2 * np.pi))
    plt.plot(x, pdf, label=f'Component {i + 1}')

plt.show()
```

```python
# Sample data composed of two normal distributions
data = np.concatenate([np.random.normal(30, 5, 5000), np.random.normal(70, 8, 500)])

# Fit a GMM to the data
gmm = GaussianMixture(n_components=2, random_state=0)
gmm.fit(data.reshape(-1, 1))

# Create a histogram of the data
plt.hist(data, bins=30, density=True, alpha=0.5, label='Data')

# Generate x-values for the PDF
x = np.linspace(min(data), max(data), 1000)

# Plot the scaled PDFs of the GMM components
for i in range(gmm.n_components):
    mean, std_dev = gmm.means_[i][0], np.sqrt(gmm.covariances_[i][0][0])
    weight = gmm.weights_[i]
    pdf = weight * np.exp(-(x - mean) ** 2 / (2 * std_dev ** 2)) / (std_dev * np.sqrt(2 * np.pi))
    plt.plot(x, pdf, label=f'Component {i + 1}')

plt.legend()
plt.title('Histogram with Overlaid GMM Components')
plt.show()
```
